```python
import pandas as pd
```

```python
data={'name':['abhinav','mangesh','ram','abhinav'],'mark':['50','85','77','89']}
df=pd.DataFrame(data)
df
```

```python
#Adding column
df['grade']=['C','A','B','A']
df
```

```python
#adding row
df.loc[4]=['sonu','65','B']
df
```

```python
#adding null column
import numpy as np
df['null_column']=np.NaN
df
```

```python
#drop column
df.drop(columns=['null_column'],inplace=True)
df
```

```python
#replace any value
df.replace(['abhinav'],'abhi',inplace=True)
df
```

```python
#drop row
df.drop([0],axis=0,inplace=True)
df
```

```python
##ading row
df.loc[5]=['abhi','55','C']
df
```

```python
#condition more then 80 and less than 99
df.query("mark > '80' and mark <'99' ")
```

```python
#change column name
df.rename(columns={'name':'firstname'},inplace=True)
df
```

```python
#name witch having A grade
df.query("grade=='A'")
```
